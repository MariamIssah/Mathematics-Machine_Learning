```python
import matplotlib.pyplot as plt

# Drop rows with missing values in relevant columns (co2, comb, and displ)
df_clean = df[['co2', 'comb', 'displ']].dropna()

# Set style
plt.style.use('ggplot')
```

```python
# YOUR CODE HERE
# Create the first subplot for CO2 Emissions Distribution
plt.figure(figsize=(18, 5))  # Set the figure size

# Histogram for CO2 Emissions Distribution
plt.subplot(1, 3, 1)
plt.hist(df_clean['co2'], bins=20, color='skyblue', edgecolor='black')  # You can adjust the number of bins
plt.title('CO2 Emissions Distribution')
plt.xlabel('CO2 (g/mile)')
plt.ylabel('Frequency')
```

```python
# Histogram: Combined Fuel Economy (MPG)
plt.figure(figsize=(8, 5))
plt.hist(df_clean['comb'], bins=10, color='seagreen', edgecolor='black')
plt.title('Combined MPG Distribution')
plt.xlabel('MPG (Combined)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Histogram: Engine Displacement (Liters)
plt.figure(figsize=(8, 5))
plt.hist(df_clean['displ'], bins=6, color='orange', edgecolor='black')
plt.title('Engine Displacement Distribution')
plt.xlabel('Displacement (Liters)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('fuel_econ.csv')

# Select only numeric columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
corr = numerical_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Fuel Economy Data")
plt.show()
```

#Interpret the data - What story does the visualization tell?
Interpretation of the Heatmap: What Story Does the Visualization Tell?

The heatmap shows how different numeric columns in the dataset are related to each other. It uses colors and numbers (called correlation values) to tell us:

🔵 Positive values (closer to +1): These columns move in the same direction (as one increases, the other increases too).

🔴 Negative values (closer to -1): These move in opposite directions (as one increases, the other decreases).

⚪ Values near 0: No clear relationship between the columns.

Now, when we look specifically at the ghgScore (Greenhouse Gas Score), we can observe:

ghgScore has a strong positive correlation with:

comb (Combined MPG)

city and highway (City and Highway MPG)

This means cars that get more miles per gallon (use less fuel) score better for greenhouse gas emissions.

🔻 ghgScore has a strong negative correlation with:

co2 (Carbon Dioxide emissions)

barrels08 (how many barrels of fuel used in a year)

fuelCost08 (how much money is spent on fuel)

That means cars that pollute more and use more fuel get lower greenhouse gas scores.
